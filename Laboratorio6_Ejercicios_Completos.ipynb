```python
# 1 Ejercicio: Algoritmo para Preparar un Té
print("=== Preparar un Té ===")
print("1. Llenar una olla con agua")
print("2. Colocar la olla en la estufa y calentar hasta que hierva")
print("3. Colocar la bolsa de té dentro de una taza limpia")
print("4. Verter el agua caliente sobre la bolsa en la taza")
print("5. Dejar reposar entre 3 y 5 minutos")
print("6. Retirar la bolsa de té de la taza")

opcion = input("¿Qué deseas agregar? (leche / azúcar / ambas / nada): ").lower()

if opcion == "leche":
    print("7. Agregar solo leche al gusto")
elif opcion == "azúcar":
    print("7. Agregar solo azúcar al gusto")
elif opcion == "ambas":
    print("7. Agregar azúcar y leche al gusto")
else:
    print("7. No se agrega ningún complemento")

print("8. Servir el té")
print(" Proceso finalizado")
```

```python
# 2 Ejercicio: Cálculo del promedio de notas
nota1 = float(input("Ingrese nota 1: "))
nota2 = float(input("Ingrese nota 2: "))
nota3 = float(input("Ingrese nota 3: "))
promedio = (nota1 + nota2 + nota3) / 3
print(f"El promedio final es: {promedio:.2f}")
```

```python
# 3 Ejercicio: Verificar si un Número es Par o Impar
numero = int(input("Ingrese un número entero: "))
if numero % 2 == 0:
    print(f"El número {numero} es PAR")
else:
    print(f"El número {numero} es IMPAR")
```

```python
# 4 Ejercicio: Bucle for para Imprimir Números del 1 al 5
print("=== Números del 1 al 5 ===")
for num in range(1, 6):
    if num % 2 == 0:
        print(f"{num} → PAR")
    else:
        print(f"{num} → IMPAR")
```

```python
# 5 Ejercicio: Convertir Grados Celsius a Fahrenheit
def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32
celsius = float(input("Ingrese grados Celsius: "))
fahrenheit = celsius_a_fahrenheit(celsius)
print(f"{celsius} °C equivalen a {fahrenheit:.2f} °F")
```

```python
# 6 Ejercicio: Contar la Cantidad de Vocales en una Cadena de Texto
texto = input("Ingrese una frase o palabra: ").lower()
vocales = "aeiou"
contador = 0
for letra in texto:
    if letra in vocales:
        contador += 1
print(f"Cantidad de vocales: {contador}")
```

```python
# 7 Ejercicio: Calcular el Factorial de un Número
n = int(input("Ingrese un número entero positivo: "))
if n < 0:
    print("El factorial no existe para números negativos")
else:
    factorial = 1
    contador = n
    while contador > 0:
        factorial *= contador
        contador -= 1
    print(f"El factorial de {n} es: {factorial}")
```
